````python
# Check for missing values
print("Missing values per column:")
print(quarters4_ytaxi.isnull().any())

# Check for duplicate rows
print("\nNumber of duplicate rows:")
print(quarters4_ytaxi.duplicated().sum())
````

````python
# Drop duplicate rows
quarters4_ytaxi.drop_duplicates(inplace=True)
print(f"\nNumber of rows after dropping duplicates: {quarters4_ytaxi.shape[0]}")

# Handle missing values in 'passenger_count' and 'RatecodeID' by filling with the mode
for col in ['passenger_count', 'RatecodeID']:
    if quarters4_ytaxi[col].isnull().any():
        mode_value = quarters4_ytaxi[col].mode()[0]
        quarters4_ytaxi[col] = quarters4_ytaxi[col].fillna(mode_value)
        print(f"Filled missing values in '{col}' with mode: {mode_value}")

# Handle missing values in 'congestion_surcharge' and 'Airport_fee' by dropping rows
cols_to_drop_na = ['congestion_surcharge', 'Airport_fee']
quarters4_ytaxi.dropna(subset=cols_to_drop_na, inplace=True)
print(f"Number of rows after dropping rows with missing values in {cols_to_drop_na}: {quarters4_ytaxi.shape[0]}")

# Verify that there are no more missing values in the handled columns
print("\nMissing values after handling:")
print(quarters4_ytaxi[['passenger_count', 'RatecodeID', 'congestion_surcharge', 'Airport_fee']].isnull().sum())
````
