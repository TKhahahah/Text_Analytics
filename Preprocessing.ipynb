### Feature enginering

````python
# Feature engineering
# extract day and time, the calculate the duration of the trip
df_quarters4_ytaxi['tpep_pickup_datetime'] = pd.to_datetime(df_quarters4_ytaxi['tpep_pickup_datetime'])
df_quarters4_ytaxi['tpep_dropoff_datetime'] = pd.to_datetime(df_quarters4_ytaxi['tpep_dropoff_datetime'])
df_quarters4_ytaxi['date'] = df_quarters4_ytaxi['tpep_pickup_datetime'].dt.normalize()
df_quarters4_ytaxi['time'] = df_quarters4_ytaxi['tpep_pickup_datetime'].dt.hour
df_quarters4_ytaxi['weekday'] = df_quarters4_ytaxi['date'].dt.day_name()
df_quarters4_ytaxi['duration'] = df_quarters4_ytaxi['tpep_dropoff_datetime'] - df_quarters4_ytaxi['tpep_pickup_datetime']
df_quarters4_ytaxi['duration'] = df_quarters4_ytaxi['duration'] / np.timedelta64(1, 'm')
````

### Handle missing and duplicates

````python
# Check missing value
df_quarters4_ytaxi.isna().sum()
````

````python
# Check duplicate row
print("Number of duplicate rows:")
print(df_quarters4_ytaxi.duplicated().sum())
````

````python
#remove duplicates
cl_quarters4_ytaxi = df_quarters4_ytaxi.drop_duplicates()
cl_quarters4_ytaxi
````

````python
#Handle missing value
cl_quarters4_ytaxi.isna().sum()
````

````python
import pandas as pd

# Columns to fill using MODE
mode_cols = ['passenger_count','RatecodeID','congestion_surcharge','Airport_fee']

# Fill missing values by MODE for selected columns
for col in mode_cols:
    mode_value = cl_quarters4_ytaxi[col].mode()[0]
    cl_quarters4_ytaxi[col] = cl_quarters4_ytaxi[col].fillna(mode_value)

print(cl_quarters4_ytaxi)
````

### Handle outliers&noise

check for outliers

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#check outliers
cols = ['passenger_count', 'trip_distance', 'fare_amount', 'tip_amount','tolls_amount','congestion_surcharge','Airport_fee']
for col in cols:
    Q1 = cl_quarters4_ytaxi[col].quantile(0.25)
    Q3 = cl_quarters4_ytaxi[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Use pandas filtering
    outliers = cl_quarters4_ytaxi[(cl_quarters4_ytaxi[col] < lower) | (cl_quarters4_ytaxi[col] > upper)]
    print(f"🔹 {col}: พบ {outliers.shape[0]} outliers") # Use .shape[0] for Pandas DataFrame row count

    # วาด boxplot ดูการกระจาย
    sns.boxplot(x=cl_quarters4_ytaxi[col])
    plt.title(f"Outliers in {col}")
    plt.show()
````

Handle outliers by filtering outliers คัดออกเฉพาะแถวที่ค่าผิดปกติชัดเจน เพื่อให้สอดคล้องกับความเป็นจริง

````python
cl_quarters4_ytaxi2 = cl_quarters4_ytaxi[
    (cl_quarters4_ytaxi['trip_distance'] > 0) & (cl_quarters4_ytaxi['trip_distance'] <= 100)
    & (cl_quarters4_ytaxi['fare_amount'] > 2.5) & (cl_quarters4_ytaxi['fare_amount'] <= 250)
    & (cl_quarters4_ytaxi['passenger_count'] >= 1) & (cl_quarters4_ytaxi['passenger_count'] < 7)
    & (cl_quarters4_ytaxi['PULocationID'] >= 1) & (cl_quarters4_ytaxi['PULocationID'] <= 263)
    & (cl_quarters4_ytaxi['DOLocationID'] >= 1) & (cl_quarters4_ytaxi['DOLocationID'] <= 263)
    & (cl_quarters4_ytaxi['duration'] > 1) & (cl_quarters4_ytaxi['duration'] <= 180)
    & (cl_quarters4_ytaxi['RatecodeID'] >= 1) & (cl_quarters4_ytaxi['RatecodeID'] <= 6)
    & (cl_quarters4_ytaxi['congestion_surcharge'] >= 0)
    & (cl_quarters4_ytaxi['Airport_fee'] >= 0)
    & (cl_quarters4_ytaxi['tip_amount'] >= 0)
    & (cl_quarters4_ytaxi['tolls_amount'] >= 0)

]
````
