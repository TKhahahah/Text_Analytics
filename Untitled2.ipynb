```python
import geopandas as gpd
import pandas as pd
import numpy as np

def clean_geo_missing(gdf: gpd.GeoDataFrame) -> gpd.GeoDataFrame:
    """
    ตรวจและจัดการ missing values สำหรับข้อมูลเชิงพื้นที่ (GeoDataFrame)
    - drop geometry ที่หาย
    - fillna สำหรับ attribute (string, numeric)
    - แสดงรายงานก่อนและหลังทำความสะอาด
    """

    print("🔍 ตรวจสอบ missing values ก่อนทำความสะอาด:")
    print(gdf.isna().sum())
    print("-" * 60)

    # 1️⃣ ลบ geometry ที่หาย (เพราะ plot/map ไม่ได้)
    if 'geometry' in gdf.columns:
        missing_geo = gdf['geometry'].isna().sum()
        if missing_geo > 0:
            print(f"🗑️ ลบ geometry ที่หาย {missing_geo} แถว")
            gdf = gdf.dropna(subset=['geometry'])

    # 2️⃣ จัดการคอลัมน์อื่น ๆ
    for col in gdf.columns:
        if col == 'geometry':
            continue  # geometry จัดการแล้ว

        if gdf[col].isna().sum() == 0:
            continue  # ไม่มี missing

        # ✅ ถ้าเป็นตัวเลข — ใช้ median (ถ้าไม่ได้ ใช้ mean)
        if pd.api.types.is_numeric_dtype(gdf[col]):
            median_val = gdf[col].median()
            if np.isnan(median_val):
                mean_val = gdf[col].mean()
                gdf[col] = gdf[col].fillna(mean_val)
                print(f"🔹 {col}: เติมค่า missing ด้วย mean = {mean_val:.2f}")
            else:
                gdf[col] = gdf[col].fillna(median_val)
                print(f"🔹 {col}: เติมค่า missing ด้วย median = {median_val:.2f}")

        # ✅ ถ้าเป็นข้อความ — เติมด้วย 'Unknown'
        elif gdf[col].dtype == 'object' or pd.api.types.is_categorical_dtype(gdf[col]):
            gdf[col] = gdf[col].fillna('Unknown')
            print(f"🔹 {col}: เติมค่า missing ด้วย 'Unknown'")

    # 3️⃣ แสดงรายงานหลังทำ
    print("\n✅ ตรวจสอบอีกครั้งหลังทำความสะอาด:")
    print(gdf.isna().sum())

    print("-" * 60)
    print(f"✅ ข้อมูลสะอาดแล้ว ({len(gdf)} แถว)")
    return gdf
```

```python
# Mapping RatecodeID to RatecodeName
ratecode_mapping = {
    1.0: 'Standard rate',
    2.0: 'JFK',
    3.0: 'Newark',
    4.0: 'Nassau/Westchester',
    5.0: 'Negotiated fare',
    6.0: 'Group ride',
    99.0: 'Unknown'
}

# Count trips for each Ratecode and Map Names
ratecode_counts_with_names = map_data['RatecodeID'].value_counts().rename(index=ratecode_mapping)

# Plot the graph
plt.figure(figsize=(10, 8))
ax = ratecode_counts_with_names.plot(
    kind='bar',
    color='#F5AD18',
    title='Ratecode Distribution (Counts)',
    ylabel='Number of Trips'
)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add trip counts on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():,}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center',
                va='bottom',
                xytext=(0, 5),
                textcoords='offset points')

# Format Y-axis as integers with commas
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.show()
```

```python
# 1️⃣ แปลงเวลารับผู้โดยสารเป็นชั่วโมง
map_data['pickup_hour'] = pd.to_datetime(
    map_data['time'],
    errors='coerce'
).dt.hour

# 2️⃣ คำนวณจำนวนทริปของแต่ละ Ratecode ต่อชั่วโมง
hourly_counts = (
    map_data.groupby(['time', 'RatecodeID'])
    .size()
    .reset_index(name='trip_count')
)

# 3️⃣ แปลงให้เป็นสัดส่วน (%) ต่อชั่วโมง
hourly_counts['share_%'] = (
    hourly_counts.groupby('time')['trip_count']
    .transform(lambda x: (x / x.sum()) * 100)
).round(2)

# 4️⃣ วาดกราฟเส้น แยกแต่ละประเภท Ratecode
plt.figure(figsize=(10,6))
for rate in hourly_counts['RatecodeID'].unique():
    sub = hourly_counts[hourly_counts['RatecodeID'] == rate]
    plt.plot(sub['time'], sub['share_%'], marker='o', label=rate)

plt.title('RatecodeID Distribution by Hour of Day')
plt.xlabel('Hour of Day (0–23)')
plt.ylabel('% of Trips')
plt.legend(title='Ratecode Type', bbox_to_anchor=(1.05,1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# เลือกข้อมูล
data = map_data['trip_distance']

# ตัด outlier 1% ขึ้นไปเพื่อให้กราฟไม่บีบแน่น
upper_limit = data.quantile(0.99)
data_clipped = data[data <= upper_limit]

plt.figure(figsize=(10, 6))

# วาด histogram + KDE
sns.histplot(
    data_clipped,
    bins=50,
    color='#9B5DE0',
    edgecolor='black',
    kde=True
)

# แสดง median
median_val = data_clipped.median()
plt.axvline(median_val, color='red', linestyle='--', label=f'Median = {median_val:.2f} miles')

# ชื่อกราฟและแกน
plt.title('Histogram of Trip Distance', fontsize=16)
plt.xlabel('Trip Distance (miles)', fontsize=12)
plt.ylabel('Number of Trips', fontsize=12)

# legend
plt.legend()

# tight layout
plt.tight_layout()
plt.show()
```

```python
# นับจำนวนครั้งที่แต่ละ PULocationID ปรากฏ
pulocation_counts = map_data['PULocationID'].value_counts().reset_index()
pulocation_counts.columns = ['PULocationID', 'Count']

# เลือกเฉพาะ LocationID ที่มีความถี่สูงสุด 20 อันดับแรก
top_pulocations = pulocation_counts.head(20)

# Merge with zone_location to get the zone names
top_pulocations = top_pulocations.merge(zone_location[['LocationID', 'Zone']], left_on='PULocationID', right_on='LocationID', how='left')

# สร้าง labels สำหรับ Treemap (แสดง Zone และจำนวนครั้ง)
labels = top_pulocations.apply(lambda x: f"{x['Zone']} ({x['Count']:,})", axis=1)

# กำหนดขนาดของแต่ละช่องใน Treemap (ตามจำนวน Count)
sizes = top_pulocations['Count'].values

# สร้างสีที่น่าสนใจสำหรับ Treemap โดยใช้ colormap 'viridis'
colors = [cm.viridis(i / float(len(labels))) for i in range(len(labels))]

# วาด Treemap
plt.figure(figsize=(14, 10))
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=.8)

# ตั้งชื่อกราฟ
plt.title('Top 20 Pickup Location Zone Treemap', fontsize=18)

# ปิดแกน x และ y เนื่องจาก Treemap ไม่ได้ใช้แกน
plt.axis('off')
plt.show()
```

```python
# นับจำนวนครั้งที่แต่ละ DOLocationID ปรากฏ
dolocation_counts = map_data['DOLocationID'].value_counts().reset_index()
dolocation_counts.columns = ['DOLocationID', 'Count']

# เลือกเฉพาะ LocationID ที่มีความถี่สูงสุด 20 อันดับแรก
top_dolocations = dolocation_counts.head(20)

# Merge with zone_location to get the zone names
top_dolocations = top_dolocations.merge(zone_location[['LocationID', 'Zone']], left_on='DOLocationID', right_on='LocationID', how='left')

# สร้าง labels สำหรับ Treemap (แสดง Zone และจำนวนครั้ง)
labels = top_dolocations.apply(lambda x: f"{x['Zone']} ({x['Count']:,})", axis=1)

# กำหนดขนาดของแต่ละช่องใน Treemap (ตามจำนวน Count)
sizes = top_dolocations['Count'].values

# สร้างสีที่น่าสนใจสำหรับ Treemap
colors = [plt.cm.plasma(i/float(len(labels))) for i in range(len(labels))] # ใช้ colormap อื่นเพื่อให้แตกต่างจาก PULocationID

# วาด Treemap
plt.figure(figsize=(14, 10))
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=.8)

# ตั้งชื่อกราฟ
plt.title('Top 20 Dropoff Location Zone Treemap', fontsize=18)

# ปิดแกน x และ y เนื่องจาก Treemap ไม่ได้ใช้แกน
plt.axis('off')
plt.show()
```

```python
# 🔹 สร้างกราฟเส้นแสดงการกระจายของ tip_amount

plt.figure(figsize=(10, 6))

# คำนวณข้อมูล Histogram ก่อน (ความถี่และขอบเขตของ bins)
# density=False หมายถึงนับจำนวน (frequency) ไม่ใช่ความหนาแน่น (density)
counts, bin_edges = np.histogram(map_data['tip_amount'], bins=50)

# คำนวณจุดกึ่งกลางของแต่ละ bin เพื่อใช้เป็นค่าบนแกน X
bin_midpoints = (bin_edges[:-1] + bin_edges[1:]) / 2.0

# วาดกราฟเส้น
plt.plot(bin_midpoints, counts, marker='o', linestyle='-', color='#1A2A4F')

# 🔹 ตั้งชื่อกราฟและแกน
plt.title('Distribution of Tip Amount')
plt.xlabel('Tip Amount ($)')
plt.ylabel('Frequency')

# 🔹 จัด layout ให้อ่านง่าย
plt.tight_layout()
plt.show()
# 🔹 นับความถี่ของ "ค่าทิป" แต่ละค่าที่ปรากฏในข้อมูล
tip_counts = map_data['tip_amount'].value_counts()

# 1. พิมพ์ 20 อันดับแรกที่พบบ่อยที่สุด (Most Frequent)
print("--- Top 20 Most Frequent Tip Amounts ---")
print(tip_counts.head(10))

print("\n" + "-"*40 + "\n")

# 2. พิมพ์ 20 อันดับแรกที่เป็นค่าสูงสุด (Highest Tip Values)
# เรา .sort_index() เพื่อเรียงตาม "ค่าทิป" (จากมากไปน้อย)
print("--- Top 20 Highest Tip Amount Values ---")
print(tip_counts.sort_index(ascending=False).head(10))
```
